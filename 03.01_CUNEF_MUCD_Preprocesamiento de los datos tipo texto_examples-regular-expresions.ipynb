```python
import re
```

```python
line = 'dividimos por espacios, sin importar si son         multiples.'
regex = re.compile('\s+')
regex.split(line)
```

```python
line = 'buscamos la palabra gato dentro de esta cadena de texto'
regex = re.compile('gato')
match = regex.search(line)
match.start(), match.end()
```

```python
line.replace('gato', 'perro')
```

```python
regex.sub('perro', line)
```

```python
line = "buscamos números como 456, 567 y 678"
```

```python
line = "buscamos un email dentro de esta cadena como paulvan@gmail.com o cunef@cunef.com"
email = re.compile('\w+@\w+\.[a-z]{3}')
email.findall(line)
```

```python
line = "eliminamos los emails dentro de esta cadena como paulvan@gmail.com"

email.sub('EMAIL', line)
```

```python
regex = re.compile(r'\$')
regex.findall("el precio son 20 \$")
```

```python
regex = re.compile('[aeiou]')
regex.split('consecutivo')
```

```python
regex = re.compile('[A-Z][0-9]')
regex.findall('1043879, G2, H6')
```

```python
regex = re.compile(r'\w{3}')
regex.findall('Los niños son majos')
```

```python
regex = re.compile(r'\w+')
regex.findall('Los niños son majos')
```

```python
email3 = re.compile(r'([\w.]+)@(\w+)\.([a-z]{3})')
```

```python
text = "Los parentesis indican grupos, que nos permiten extraer entidades específicas como usuario, dominio y subdominio: paul@cunef.com, cunef@cunef.com"
email3.findall(text)
```

```python
email4 = re.compile(r'(?P<user>[\w.]+)@(?P<domain>\w+)\.(?P<suffix>[a-z]{3})')
match = email4.match('paul@cunef.com')
match.groupdict()
```
